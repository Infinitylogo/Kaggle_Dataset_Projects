# Performing cleaning, tokenization, steming

```python
stemmer = PorterStemmer() 
tokenizer = RegexpTokenizer(r'[a-zA-Z]+')
```

```python
from nltk.corpus import stopwords
from collections import Counter
```

# Tokenization

```python
def data_tokenization(text):
  #tokenization of words such as "unit", "citizen"
  data_tokens=tokenizer.tokenize(text) #iterating over each texts
  return data_tokens #returning tokens 
```

# Performing Steaming

```python
def Steaming_tokens(tokenized_values, stop_words):
  stem_words=[]
  for w in tokenized_values:
    if w not in stop_words:
      stem_words.append(stemmer.stem(w))
  return stem_words
```

# Converting each words into lower case.

```python
def lower_case_conversion(text_corpous):
   return text_corpous.lower()
```

# Combining all the function to get the spefic texts for getting the results

```python
from math import sqrt #math function for calculating square root
import math
from math import log10 #for logirithmic calculations
```

```python
#Data corpous
text_data='/content/drive/MyDrive/data_text/US_Inaugural_Addresses/US_Inaugural_Addresses'
```

```python
TokenFrequency=Counter()  
#dict for tokens as value and docs name as key                      
Corpous_Dict={}                              
for Document_Name in os.listdir(text_data):
    #loading data from all text files
    corpous_data = open(os.path.join(text_data, Document_Name), "r", encoding='windows-1252')
    #reading data from each files 
    Docs_Corpous = corpous_data.read()
    corpous_data.close() #closing 
    #converting data to lower case..
    Docs_Corpous = lower_case_conversion(Docs_Corpous)
    #perofrming tokenization and calling above function for getting tokens         
    Corpous_Token = data_tokenization(Docs_Corpous) 
    #stop words                  
    stop_word=stopwords.words('english')
    #Root words 
    Corpous_Token = Steaming_tokens(Corpous_Token, stop_word)           
    term_freq=Counter(Corpous_Token) #total count of frequency the words comes in entire documents 
    TokenFrequency+=Counter(list(set(Corpous_Token)))  #total frequency of tokens      
    Corpous_Dict[Document_Name]=term_freq.copy() #creating dic having documents and values with frequency and tokens                                         
    term_freq.clear()                                                       
```

# Fucntion to calculate term frequency....

```python
def getidf(DataTokens):
    #Calculate IDF 
    if TokenFrequency[DataTokens]==0:
        return -1
    return log10(len(Corpous_Dict)/TokenFrequency[DataTokens]) 
```

# Function to calcualte weights....

```python
def calWeight(Document_Name, DataTokens):  
    #IDF FREQUENCY
    Idf_Freq=getidf(DataTokens)  # getting freq of data 
    return (1+log10(Corpous_Dict[Document_Name][DataTokens]))*Idf_Freq  
```

# Function to get the query as input and print the results having *TXT* and *Similarity score*...

```python
#DICT INITIALIZATION
Vector_Dict={}                                                                
VectorDocList={}  
```

# CHECKING LENGTH OF DATA TO PERFORM QUERY ON TEXTS TO GET SIMILARITY SCORE

```python
def Checking_Length(text):
  text=text.lower() 
  #LOADING TEXT IN LOWER CASE                    
  VecTransform,LocationVecDocs,CommonValues={},{},{} #INITIALIZE DICT 
  VecLength=FlagValue=0 # LENGTH AND FLAG VALUES 
  Cosimilarity_Score=Counter() # SIMILARITY SCORE 
  Cosine_Similarity_Score=set()                         
  for DataTokens in text.split(): 
      DataTokens=stemmer.stem(DataTokens)  
      if DataTokens not in VectorDocList:          
          continue
      if getidf(DataTokens)==0:            
          LocationVecDocs[DataTokens], weights = zip(*VectorDocList[DataTokens].most_common())         
      else:
          LocationVecDocs[DataTokens],weights = zip(*VectorDocList[DataTokens].most_common(20)) 
      # print(DataTokens)
      if len(weights)>=1:
        CommonValues[DataTokens]=weights[-1]
      else:
        CommonValues[DataTokens]=weights[0]                                                        
      if FlagValue==1:
          Cosine_Similarity_Score=set(LocationVecDocs[DataTokens]) & Cosine_Similarity_Score    #  Cosine_Similarity_Score                           
      else:
          Cosine_Similarity_Score=set(LocationVecDocs[DataTokens])
          FlagValue=1
      Text_Count=text.count(DataTokens) 
      if Text_Count!=0:
        VecTransform[DataTokens]=1+log10(Text_Count)
      else:
        VecTransform[DataTokens]=1
      VecLength+=VecTransform[DataTokens]**2                      
  VecLength=sqrt(VecLength)
  return VecTransform, LocationVecDocs, VecLength, CommonValues, Cosimilarity_Score,Cosine_Similarity_Score
```

# For Weights 

```python
def CheckingWeight(VecTransform, LocationVecDocs, VecLength, CommonValues, Cosimilarity_Score, max, Results=0):
  for doc in Vector_Dict:
      cosine_similarity=0
      for token in VecTransform:
          if doc in LocationVecDocs[token]:
              cosine_similarity = cosine_similarity + (VecTransform[token] / VecLength) * VectorDocList[token][doc]      
          else:
              cosine_similarity = cosine_similarity + (VecTransform[token] / VecLength) * CommonValues[token]                    
      Cosimilarity_Score[doc]=cosine_similarity
  max=Cosimilarity_Score.most_common(1)                                                              
  Results,weight=zip(*max)
  return Results, weight
```

# Checking Queries

```python
def query(text):                             
    VecTransform, LocationVecDocs, VecLength, CommonValues, Cosimilarity_Score,Cosine_Similarity_Score=Checking_Length(text)
    Results, weight=CheckingWeight(VecTransform, LocationVecDocs, VecLength, CommonValues, Cosimilarity_Score, max, Results=0)
    try:
        if Results[0] in Cosine_Similarity_Score:                                                             
            return Results[0],weight[0]
        else:
            return "Find More",0                                                            
    except UnboundLocalError:                                                                
        return "Not found",0
```

# Calculate tf-idf and document length

```python
DocumentSize=Counter()
```

```python
for Document_Name in Corpous_Dict:
    Vector_Dict[Document_Name]=Counter()                    
    length=0
    for DataTokens in Corpous_Dict[Document_Name]:
        weight = calWeight(Document_Name, DataTokens)             
        Vector_Dict[Document_Name][DataTokens]=weight  
        length += weight**2                                    
    DocumentSize[Document_Name]=math.sqrt(length)
```

# Normalize Token_Weights

```python
for Document_Name in Vector_Dict:
    for DataTokens in Vector_Dict[Document_Name]:
        Vector_Dict[Document_Name][DataTokens]= Vector_Dict[Document_Name][DataTokens] / DocumentSize[Document_Name]   # getting name and tokens in dict
        if DataTokens not in VectorDocList: 
            VectorDocList[DataTokens]=Counter()  
        VectorDocList[DataTokens][Document_Name]=Vector_Dict[Document_Name][DataTokens] 
```
